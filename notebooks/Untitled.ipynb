```python
a.obs.index = [x.replace('-', '.') for x in a.obs.index]
a.var.index = [x.split('.')[0] for x in a.var.index]
a.var.index = [x.replace('-', '_') for x in a.var.index]
```

```python
from collections import Counter
for dataset in set(a.obs['dataset']):
    print(dataset)
    for t, c in Counter(a[a.obs['dataset']==dataset].obs['final_cell_types']).most_common():
        print(t, c)
```

```python
# only take in-house for reference
a = a[[True if 'In-house' in d else False for d in a.obs['dataset']]]
a
```

```python
new = a.copy()
new.X = a.layers['counts']
```

```python
sorted(set(new.obs['cell_type']))
```

```python
# remove the high MT cluster
new = new[new.obs['cell_type_level_3']!='U1']
new
```

```python
set(new.X[0].toarray().flatten())
```

```python
# make R friendly
m = {
    'CD8 T cell - exausted': 'CD8_T_cell',
    'Malignant_0': 'Malignant',
    'Malignant_1': 'Malignant',
    'Malignant_2': 'Malignant',
    'Malignant_3': 'Malignant',
    'Malignant_4': 'Malignant',
    'Other': 'Stellate',
}
new.obs['cell_type_level_3'] = [m.get(x, x) for x in new.obs['cell_type_level_3']]
new.obs['cell_type_level_3'] = [x.replace(' ', '_') for x in new.obs['cell_type_level_3']]
sorted(set(new.obs['cell_type_level_3']))
```

```python
df = pd.DataFrame(new.X.transpose().toarray(), index=new.var.index, columns=new.obs['cell_type_level_3'].to_list())
df
```

```python
Counter(df.columns).most_common()
```

```python
df.index.name = 'Gene'
df
```

```python
df = df.groupby('Gene').mean()
df
```
