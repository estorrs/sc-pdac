```r
df = read.csv('../data/cibersortx/bailey.txt', sep='\t')
rownames(df) <- df$Gene
df$Gene = NULL
bailey = CreateSeuratObject(df, project = 'bailey', assay = 'RNA', names.field = 'Gene')
bailey
```

```r
df = read.csv('../data/cibersortx/tcga.txt', sep='\t')
rownames(df) <- df$Gene
df$Gene = NULL
tcga = CreateSeuratObject(df, project = 'tcga', assay = 'RNA', names.field = 'Gene')
tcga
```

```r
df = read.csv('../data/cibersortx/moffit.txt', sep='\t')
rownames(df) <- df$Gene
df$Gene = NULL
moffit = CreateSeuratObject(df, project = 'moffit', assay = 'RNA', names.field = 'Gene')
moffit
```

```r
df = read.csv('../data/cibersortx/kirby.txt', sep='\t')
rownames(df) <- df$Gene
df$Gene = NULL
kirby = CreateSeuratObject(df, project = 'kirby', assay = 'RNA', names.field = 'Gene')
kirby
```

```r
pancreas.list <- c(tcga, bailey, moffit, kirby)
for (i in 1:length(pancreas.list)) {
    obj = pancreas.list[[i]]
    pancreas.list[[i]] <- SCTransform(obj, verbose = FALSE)
}
```

```r
pancreas.features <- SelectIntegrationFeatures(object.list = pancreas.list, nfeatures = 3000)
```

```r
pancreas.list <- PrepSCTIntegration(object.list = pancreas.list, anchor.features = pancreas.features, 
    verbose = FALSE)
```

```r
pancreas.anchors <- FindIntegrationAnchors(object.list = pancreas.list, normalization.method = "SCT", 
    anchor.features = pancreas.features, verbose = FALSE)
pancreas.integrated <- IntegrateData(anchorset = pancreas.anchors, normalization.method = "SCT", 
    verbose = FALSE)
```

```r
pancreas.integrated <- RunPCA(pancreas.integrated, verbose = FALSE)
pancreas.integrated <- RunUMAP(pancreas.integrated, dims = 1:30)
```
